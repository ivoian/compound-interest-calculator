```python
def compound_interest(principal, rate, time, contribution=0):
    """
    principal: initial investment (float)
    rate: interest rate per period (decimal, e.g., 0.05 = 5%)
    time: number of periods (int)
    contribution: additional contribution each period (float)
    """
    balance = principal
    history = []

    for t in range(1, time + 1):
        balance = (balance + contribution) * (1 + rate)
        history.append((t, round(balance, 2)))

    return balance, history


# Example usage
initial_investment = 1000   # $1000
interest_rate = 0.01        # 1% per month
periods = 12                # 12 months
monthly_contribution = 100  # $100 each month

final_balance, growth = compound_interest(initial_investment, interest_rate, periods, monthly_contribution)

print(f"Final Balance after {periods} periods: ${final_balance:.2f}\n")

print("Investment Growth:")
for period, value in growth:
    print(f"Month {period}: ${value}")
```

```python
import matplotlib.pyplot as plt

# Extracting data for the chart
months = [p for p, v in growth]
values = [v for p, v in growth]

plt.figure(figsize=(9,5))
plt.plot(months, values, marker='o', linestyle='-', color='royalblue')

plt.title("Compound Interest Growth", fontsize=14, fontweight="bold")
plt.xlabel("Months")
plt.ylabel("Balance ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.fill_between(months, values, color="skyblue", alpha=0.3)

plt.show()
```
